```python
#import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#load dataset

epl_df = pd.read_csv('C:\\3) Learning\\Python - Numpy Pandas\\EPL Analysis\\EPL_20_21.csv')
epl_df.head()
```

```python
#normal analysis
epl_df.info()
```

```python
#detailed analysis
epl_df.describe()
```

```python
#checking null values
epl_df.isna().sum()
```

```python
#create 2 Columns
epl_df['MinsPerMatch'] = (epl_df['Mins']/epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch'] = epl_df['Goals']/epl_df['Matches'].astype(float)
epl_df.head()
```

```python
#total goals
Total_goals = epl_df['Goals'].sum()
print(Total_goals)
```

```python
#Penalty goals
total_PenaltyGoals = epl_df['Penalty_Goals'].sum()
print(total_PenaltyGoals)
```

```python
#penalties attemptes
total_penaltiesAttempt = epl_df['Penalty_Attempted'].sum()
print(total_penaltiesAttempt)
```

```python
#chart for penalties missed vs scored
plt.figure(figsize =(13,6))
plt_not_scored = epl_df['Penalty_Attempted'].sum() - total_PenaltyGoals
data =[plt_not_scored,total_PenaltyGoals]
labels = ['Penalies missed','Penalies Scored']
color = sns.color_palette('Set2')
plt.pie(data,labels = labels ,colors = color, autopct='%.0f%%')
plt.show()
```

```python
#unique positions
epl_df['Position'].unique()
```

```python
#total FW players
epl_df[epl_df['Position'] == 'FW']
```

```python
#players fro different nations
np.size((epl_df['Nationality'].unique()))
```

```python
#top 10 nations players from country 
nationality =  epl_df.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind='bar', figsize=(12,6), color=sns.color_palette("magma"))

```

```python
#clubs with max players in squad

epl_df['Club'].value_counts().nlargest(5).plot(kind='bar',color = sns.color_palette("viridis"))
```

```python
#clubs with the least players in squad
epl_df['Club'].value_counts().nsmallest(6).plot(kind = 'bar',color = sns.color_palette("viridis"))
```

```python
#players based on ages
Under20 = epl_df[epl_df['Age'] <= 20]
Age20_25 = epl_df[(epl_df['Age'] >20) & (epl_df['Age'] <= 25)]
Age25_30 = epl_df[(epl_df['Age'] >25) & (epl_df['Age'] <= 30)]
Above30 = epl_df[epl_df['Age'] > 30]
```

```python
#graph as per above categorization
x = np.array([Under20['Name'].count(),Age20_25['Name'].count(),Age25_30['Name'].count(),Above30['Name'].count()])
mylabels = ["<20" , ">20 & <25" , ">25 & < 30" , ">30"]
plt.title('Total Players with Age groups',fontsize = 20)
plt.pie(x, labels = mylabels , autopct= "%.1f%%")
plt.show()
```

```python
#total under 20 players for each club

player_under_20 = epl_df[epl_df['Age'] < 20]
player_under_20['Club'].value_counts().plot(kind = 'bar', color = sns.color_palette ("cubehelix"))
```

```python
#under 20 players in Manu
player_under_20[player_under_20["Club"] == 'Manchester United']
```

```python
#under 20 players in Chelsea
player_under_20[player_under_20["Club"] == 'Chelsea']
```

```python
#Average Age of players
plt.figure(figsize=(12,6))
sns.boxplot(x ='Club', y = 'Age', data = epl_df)
plt.xticks(rotation = 90)
```

```python
#verify 
num_players = epl_df.groupby('Club').size()
data = (epl_df.groupby('Club')['Age'].sum())/num_players
data.sort_values(ascending=False)
```

```python
#total Assist from each Club
Assist_by_Club = pd.DataFrame(epl_df.groupby('Club',as_index =False)['Assists'].sum())
sns.set_theme(style ="whitegrid", color_codes = True)
ax = sns.barplot (x = 'Club', y = 'Assists', data = Assist_by_Club.sort_values(by="Assists"), palette= 'Set2')
ax.set_xlabel ("Club", fontsize = 25)
ax.set_ylabel ("Assists", fontsize = 20)
plt.xticks(rotation = 75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Clubs vs Total Assists', fontsize = 20)
```

```python
#top 10 assists

top_10_assists =  epl_df[['Name','Club','Assists','Matches']].nlargest(n=10, columns ='Assists')
top_10_assists
```

```python
#goals by clubs
Goals_by_Club = pd.DataFrame(epl_df.groupby('Club',as_index =False)['Goals'].sum())
sns.set_theme(style ="whitegrid", color_codes = True)
ax = sns.barplot (x = 'Club', y = 'Goals', data = Goals_by_Club.sort_values(by="Goals"), palette= 'rocket')
ax.set_xlabel ("Club", fontsize = 25)
ax.set_ylabel ("Goals", fontsize = 20)
plt.xticks(rotation = 75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Clubs vs Total Goals', fontsize = 20)
```

```python
#most goals by player
top_10_goals = epl_df[['Name','Club','Goals','Matches']].nlargest(n=10, columns='Goals')
top_10_goals
```

```python
#Goals per Match
top_10_goals_per_match = epl_df[['Name','GoalsPerMatch','Matches','Goals']].nlargest(n=10, columns='GoalsPerMatch')
top_10_goals_per_match
```
